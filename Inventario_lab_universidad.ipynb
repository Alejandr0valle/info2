```python
class Sustancia:
    def __init__(self, serial, fila, columna, libre_acceso, peligrosa):
        self.serial = serial
        self.fila = fila
        self.columna = columna
        self.libre_acceso = libre_acceso
        self.peligrosa = peligrosa

class AcidoBase(Sustancia):
    def __init__(self, serial, fila, columna, libre_acceso, peligrosa, concentracion, efectos_salud):
        super().__init__(serial, fila, columna, libre_acceso, peligrosa)
        self.concentracion = concentracion
        self.efectos_salud = efectos_salud

class Alcohol(Sustancia):
    def __init__(self, serial, fila, columna, libre_acceso, peligrosa, temp_ebullicion, temp_arder):
        super().__init__(serial, fila, columna, libre_acceso, peligrosa)
        self.temp_ebullicion = temp_ebullicion
        self.temp_arder = temp_arder

class Solvente(Sustancia):
    def __init__(self, serial, fila, columna, libre_acceso, peligrosa, naturaleza, viscosidad):
        super().__init__(serial, fila, columna, libre_acceso, peligrosa)
        self.naturaleza = naturaleza
        self.viscosidad = viscosidad

class Otro(Sustancia):
    def __init__(self, serial, fila, columna, libre_acceso, peligrosa, fecha_ingreso, motivo_ingreso, cantidad):
        super().__init__(serial, fila, columna, libre_acceso, peligrosa)
        self.fecha_ingreso = fecha_ingreso
        self.motivo_ingreso = motivo_ingreso
        self.cantidad = cantidad

class Inventario:
    def __init__(self):
        self.inventario = {}
        
    def agregar_sustancia(self, sustancia):
        if sustancia.serial in self.inventario:
            print("Error: la sustancia con serial {} ya existe en el inventario".format(sustancia.serial))
        else:
            self.inventario[sustancia.serial] = sustancia
            print("Sustancia con serial {} agregada al inventario".format(sustancia.serial))
            
    def actualizar_sustancia(self, serial, nueva_sustancia):
        if serial in self.inventario:
            self.inventario[serial] = nueva_sustancia
            print("Sustancia con serial {} actualizada en el inventario".format(serial))
        else:
            print("Error: la sustancia con serial {} no existe en el inventario".format(serial))
            
    def borrar_sustancia(self, serial):
        if serial in self.inventario:
            del self.inventario[serial]
            print("Sustancia con serial {} eliminada del inventario".format(serial))
        else:
            print("Error: la sustancia con serial {} no existe en el inventario".format(serial))
            
    def mostrar_inventario(self):
        print("Inventario:")
        for serial, sustancia in self.inventario.items():
            print("Serial: {}".format(serial))
            if isinstance(sustancia, AcidoBase):
                print("Tipo: Ácido/Base")
                print("Concentración: {} M".format(sustancia.concentracion))
                print("Efectos sobre la salud: {}".format(sustancia.efectos_salud))
            elif isinstance(sustancia, Alcohol):
                print("Tipo: Alcohol")
                print("Temperatura de ebullición: {}°C".format(sustancia.temp_ebullicion))
                print("Temperatura de ardor: {}°C".format(sustancia.temp_arder))
            elif isinstance(sustancia, Solvente):
                print("Tipo: Solvente")
                print("Naturaleza: {}".format(sustancia.naturaleza))
                print("Viscosidad: {}".format(sustancia.viscosidad))
            elif isinstance(sustancia, Otro):
                print("Tipo: Otro")
                print("Fecha de ingreso: {}".format(sustancia.fecha_ingreso))
                print("Motivo de ingreso: {}".format(sustancia.motivo_ingreso))
                print("Cantidad: {}".format(sustancia.cantidad))
            else:
                print("Tipo: Sustancia")
            print("Fila: {}".format(sustancia.fila))
            print("Columna: {}".format(sustancia.columna))
            print("Libre acceso: {}".format(sustancia.libre_acceso))
            print("Peligrosa: {}".format(sustancia.peligrosa))
            print("-----------")  

   

inventario = Inventario()

#menú de opciones
def mostrar_menu():
    print("1. Agregar una sustancia al inventario")
    print("2. Actualizar una sustancia existente en el inventario")
    print("3. Borrar una sustancia del inventario")
    print("4. Mostrar todo el inventario")
    print("0. Salir")

# las opciones del usuario
while True:
    # Mostrar el menú
    mostrar_menu()

    # Leer la opción del usuario
    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        # Agregar una sustancia al inventario
        while True:
            try:
                codigo1 = int(input("Ingrese el código de la sustancia: "))
                nombre1 = input("Ingrese el nombre de la sustancia: ")
                tipo1 = input("Ingrese el tipo de la sustancia (A para ácido, B para alcohol, C para solvente): ").upper()
                if tipo1 == 'A':
                    acido1 = AcidoBase(codigo1, nombre1, 1, False, True, 0.5, 'Irritación de ojos')
                    inventario.agregar_sustancia(acido1)
                elif tipo1 == 'B':
                    contenido1 = float(input("Ingrese el contenido de la sustancia: "))
                    alcohol1 = Alcohol(codigo1, nombre1, 2, True, False, contenido1, 12)
                    inventario.agregar_sustancia(alcohol1)
                elif tipo1 == 'C':
                    polaridad1 = input("Ingrese la polaridad de la sustancia (Polar o No polar): ")
                    punto_ebullicion1 = float(input("Ingrese el punto de ebullición de la sustancia: "))
                    solvente1 = Solvente(codigo1, nombre1, 3, False, False, polaridad1, punto_ebullicion1)
                    inventario.agregar_sustancia(solvente1)
                else:
                    print("Tipo de sustancia no válido.")
                    continue
                break
            except ValueError:
                print("Ingrese una entrada válida.")

    elif opcion == "2":
        # Actualizar una sustancia existente en el inventario
        while True:
            try:
                codigo2 = int(input("Ingrese el código de la sustancia que desea actualizar: "))
                nuevo_nombre2 = input("Ingrese el nuevo nombre de la sustancia: ")
                nuevo_tipo2 = input("Ingrese el nuevo tipo de la sustancia (A para ácido, B para alcohol, C para solvente): ").upper()
                if nuevo_tipo2 == 'A':
                    nuevo_acido2 = AcidoBase(codigo2, nuevo_nombre2, 4, True, True, 1.0, 'Corrosión de piel')
                    inventario.actualizar_sustancia(codigo2, nuevo_acido2)
                elif nuevo_tipo2 == 'B':
                    nuevo_contenido2 = float(input("Ingrese el nuevo contenido de la sustancia: "))
                    nuevo_alcohol2 = Alcohol(codigo2, nuevo_nombre2, 4, True, False, nuevo_contenido2, 12)
                    inventario.actualizar_sustancia(codigo2, nuevo_alcohol2)
                elif nuevo_tipo2 == 'C':
                    nuevo_polaridad2 = input("Ingrese la nueva polaridad de la sustancia (Polar o No polar): ")
                    nuevo_punto_ebullicion2 = float(input("Ingrese el nuevo punto de ebullición de la sustancia: "))
                    nuevo_solvente2 = Solvente(codigo2, nuevo_nombre2, 4, False, False, nuevo_polaridad2, nuevo_punto_ebullicion2)
                    inventario.actualizar_sustancia(codigo2, nuevo_solvente2)
                else:
                    print("Tipo de sustancia no válido.")
                    continue
                break
            except ValueError:
                print("Ingrese una entrada válida.")

    elif opcion == "3":
        # Borrar una sustancia del inventario
        codigo3 = input("Ingrese el código de la sustancia que desea borrar: ")
        inventario.borrar_sustancia(codigo3)

    elif opcion == "4":
        # Mostrar todo el inventario
        inventario.mostrar_inventario()

    elif opcion == "0":
        # Salir del programa
        break

    else:
        print("Opción no válida. Intente de nuevo.")
```
